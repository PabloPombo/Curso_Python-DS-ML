```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Cursos UDEMY\Curso_Python-DS-ML\Curso_Python-DS-ML\Día6-Pandas-p1\medallas.csv")
df.info()
```

```python
df = df.fillna(value=0)
df
```

```python
df[["Oro","Plata","Bronce"]] = df[["Oro","Plata","Bronce"]].apply(lambda x: x.astype(int))
df
```

```python
df
```

```python
df_sortoro = df.sort_values(by="Oro",ascending=False, ignore_index=True)
print("3 Paises con mas medallas de Oro")
df_sortoro.head(3)
```

```python
df_sorttotal = df.sort_values(by="Total",ascending=False, ignore_index=True)
df_sorttotal = df_sorttotal[df_sorttotal["Total"] > 10]
df_sorttotal
```
