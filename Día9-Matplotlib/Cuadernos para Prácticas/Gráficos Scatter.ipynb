```python
import matplotlib.pyplot as plt
import numpy as np
# Datos para las series del gráfico
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.scatter(x1,y1,color= 'blue', marker='o',label='Serie 1')
plt.scatter(x2,y2,color= 'green', marker='^',label='Serie 2')
plt.legend()
```
