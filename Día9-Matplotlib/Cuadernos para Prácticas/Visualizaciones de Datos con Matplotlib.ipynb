```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,11)
y = x**2

plt.stem(x,y)
plt.title("Relación entre Números y sus Cuadrados")
plt.xlabel("Número (x)")
plt.ylabel("Cuadrado del número (x^2)")
```

```python
import matplotlib.pyplot as plt  # Importar la biblioteca de gráficos Matplotlib
 
# Datos de entrada
x = range(1, 11)  # Crear una lista de números enteros del 1 al 10
y = [i**2 for i in x]  # Calcular el cuadrado de cada número en la lista x
 
# Crear gráfico de tallo
fig, ax = plt.subplots(1,2)  # Inicializar una figura y un conjunto de subgráficos
ax[0].stem(x, y)  # Crear un gráfico de tallo para mostrar los datos de x e y
 
# Añadir títulos y etiquetas
ax[0].set_title("Relación entre Números y sus Cuadrados")  # Establecer el título del gráfico
ax[0].set_xlabel("Número (x)")  # Establecer la etiqueta del eje x
ax[0].set_ylabel("Cuadrado del número (x^2)")  # Establecer la etiqueta del eje y
#plt.title("Graficos")
```

```python
import matplotlib.pyplot as plt

y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
x = range(1,6)

plt.style.use("classic")
fig, ax = plt.subplots()

ax.stackplot(x, [y1,y2,y3], colors=(['red','blue','green']), labels=(['Serie y1','Serie y2','Serie y2']))
ax.legend(loc='upper left')
ax.set_title("Distribución acumulativa de valores por serie de tiempo")
ax.set_xlabel("Tiempo")
ax.set_ylabel("Valores Acumulados")
```

```python
import matplotlib.pyplot as plt
 
# Definir los datos
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
x = range(1, 6)  # Tiempos del 1 al 5
 
# Crear el gráfico apilado
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, colors=['red', 'blue', 'green'], labels=['Serie y1', 'Serie y2', 'Serie y3'])
 
# Añadir leyenda
ax.legend(loc='upper left')
 
# Añadir títulos y etiquetas
ax.set_title('Distribución acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0) # Establecer la misma semilla para la generación de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)

fig, ax = plt.subplots()

ax.violinplot((data1,data2,data3), data=(['Muestra1','Muestra2','Muestra3']))

ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])

ax.set_title("Comparación de distribuciones en Violin Plot")
```
